#서울시 생활인구 데이터를 가공하겠습니다.
전기차를 이용이 적은 연령대를 삭제하고, 시간대별 행정동 인구수의 평균을 취하고자 합니다.

```python
import pandas as pd

df = pd.read_excel('code_to_name_LPdata(2).xlsx')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df2 = df.groupby(['행정동명','시간대구분'])
```

```python
df['총생활인구수2'] = df.iloc[:,-22:].sum(axis=1)
# 기존에 총생활인구수는 10대 남성 여성도 포함되어있는 수입니다. 
# code_to_name_LPdata를 만들 당시에 10대 인구를 빼놨습니다.
# 그래서 10대를 제외한 새로운 총생활인구2를 만들었습니다.
# 기존 총생활인구수는 10대 포함입니다.
```

```python
df.head()
```

```python
# 필요한 column만 따로 모았습니다.
df_pop = df[['기준일ID','시간대구분','행정동명','총생활인구수2']]
df_pop.head()
```

```python
df_pop.rename(columns={'총생활인구수2':'총생활인구'})
```

```python
df = df_pop
```

```python
df.head()
```

```python
df = df.rename(columns={'총생활인구수2':'총생활인구수'})
```

# 행정동명을 기준으로 같은 시간대에의 총생활인구수의 평균을 구해줍니다.

```python

df2 = df.groupby(['행정동명','시간대구분']).mean().drop(columns='기준일ID')
```

```python
df2 = df2.rename(columns={'총생활인구수2':'총생활인구수'})
df2
```

```python
df3 = df.groupby('행정동명').mean().drop(columns='기준일ID')
df3
```

```python
df3.info
```

# 위방법을 사용하여 시간을 0~9시, 9시~18시, 18시~24시 구분해서 만들어봅시다.


```python

zero2nine = df.query("시간대구분<=9")
nine2eighteen = df.query("시간대구분>=9 and 시간대구분<=18")
eighteen224 = df.query("시간대구분>=18")
```

```python
zero2nine
# 0~9시 사이에 데이터가 잘 추출되었네요!
```

```python
df_zero2nine = zero2nine.groupby('행정동명').mean().drop(columns=['기준일ID','시간대구분'])
df_zero2nine
```

```python
df_nine2eighteen = nine2eighteen.groupby('행정동명').mean().drop(columns=['기준일ID','시간대구분'])
df_eighteen224 = eighteen224.groupby('행정동명').mean().drop(columns=['기준일ID','시간대구분'])
```

```python
df_zero224 = df3
df_zero224.drop(columns='시간대구분')
```

```python
df_zero2nine.to_excel('zero2nine.xlsx')
df_nine2eighteen.to_excel('nine2eighteen.xlsx')
df_eighteen224.to_excel('eighteen224.xlsx')
df_zero224.to_excel('zero224.xlsx')
```
